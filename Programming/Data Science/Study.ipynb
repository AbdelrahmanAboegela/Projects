```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency='eur', days=30)
```

```python
bitcoin_data
```

```python
timestamps = [timestamp[0] / 1000 for timestamp in bitcoin_data['prices']]
prices = [price[1] for price in bitcoin_data['prices']]


data = pd.DataFrame(list(zip(timestamps, prices)), columns=['TimeStamp', 'Price'])


print(data)
```

```python
data['Date'] = pd.to_datetime(data['TimeStamp'],unit ='ms')
```

```python
print(data)
```

```python
data.plot.hist()
```
